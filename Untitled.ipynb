### 2 . Preprocess

#### 2.1 Scaling the images

```python
# The data is somehow preprocessed already as the dataset utility helpd used resize the images but lets make sure they are all same size
```

```python
# the easiest way to normalize a picture is dividing it by 255
#data is our previously defined variable and lambda is just an action occurring in the same line in this case x divided by 255
# x is our pictures and Y is our label (NORMAL/PNEUMONIA)
data = data.map(lambda x,y: (x/255, y))
```

#### 2.2 Split Data

```python
# first check how many data do we have
len(data)
```

```python
#we will divide our data into 70% to train 20% to validate and 10% to test
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)+1 # added one so that we use all the data
test_size = int(len(data)*.1)
```

```python
#if for example  train size is 7 as the len of our data is a total 10 , the data len is how many batches do we have
#remember that each batch contains 32 images so we can say that we have a maximum of (10 batches times 32 images per batch )=320 images
# althogh in real life we only have 304 that is because if we multiply 9*32 it is 288 
#the misising 16 images will be allocated in another batch doesnt matter if it fill it up or not

train_size
```

```python
#check the size of validation
val_size
```

```python
#check the size of test
test_size
```

```python
#check the total size , if it does not match with the len of data some modification must be don on the splitting
train_size+val_size+test_size
```

```python
#here we devide the data we know already how much data we need
train = data.take(train_size)
#here we define to exclude the train that was already taken
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

### 3 Creating the model!

#### 3.1 Build the architecture

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense,Flatten,Dropout, MaxPooling2D
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
```

```python
model.add(Conv2D(50, (3,3), 1, activation='relu', input_shape=(300,300,3)))
model.add(MaxPooling2D(2,2))


model.add(Conv2D(25, (3,3), 1, activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.5))

model.add(Conv2D(15, (3,3), 1, activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.5))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
#####################################################################################
```

```python
model.compile(Adam(learning_rate=0.001), loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

#### 3.2 TRAIN!!!!!

```python
logdir='logs'
```

```python
%load_ext tensorboard
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
tic = time.perf_counter()
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
hist = model.fit(train, epochs=10, validation_data=val, callbacks=[tensorboard_callback])
toc = time.perf_counter()
print(f"Training took {toc - tic:0.4f} seconds")
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

### 4. Evaluate

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(pre.result(), re.result(), acc.result())
```
